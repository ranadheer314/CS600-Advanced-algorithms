```python
import random 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1=np.random.uniform(0,2,100)
y_1=np.random.uniform(0,2,100)
x_2=np.random.uniform(0,2,1000)
y_2=np.random.uniform(0,2,1000)
x_3=np.random.uniform(0,2,100000)
y_3=np.random.uniform(0,2,100000)
m1=100
m2=1000
m3=100000
```

```python
def number_of_grey1(x,y):
    count=0
    for i in range(x.shape[0]):
        if (((x[i]-1)**2 + (y[i]-1)**2) <=1) and ((x[i])**2 + ((y[i])**2) >=4):
            count=count +1
    return count
```

```python
def number_of_grey2(x,y):
    count=0
    for i in range(x.shape[0]):
        if (((x[i]-2)**2 + (y[i]-2)**2) >=4) and ((x[i]-1)**2 + ((y[i]-1)**2) <=1):
            count=count +1
    return count
```

```python
n1=number_of_grey1(x_1,y_1)
n2=number_of_grey2(x_1,y_1)

n10=number_of_grey1(x_2,y_2)
n20=number_of_grey2(x_2,y_2)

n100=number_of_grey1(x_3,y_3)
n200=number_of_grey2(x_3,y_3)
```

```python
t1=n1+n2
t2=n10+n20
t3=n100+n200
total_area_m1=4*t1/m1

total_area_m2=4*t2/m2

total_area_m3=4*t3/m3
```
